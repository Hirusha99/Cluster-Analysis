## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set

from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
data = pd.read_csv('/content/005 Categorical.csv')
data.head()
```

## Map the data

Use the <i>'continent'</i> category for this analysis.

```python
data['continent'].unique()
```

```python
data['continent'] = data['continent'].map({'North America':0,'Asia':1,'Africa':2,'Europe':3,'South America':4,'Oceania':5,'Antarctica':6,'Seven seas (open ocean)':7})
data.head()
```

## Select the features

```python
x = data.iloc[:,3:4]
x
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(4)
```

## Clustering results

```python
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

## Plot the data

```python
plt.scatter(data['Longitude'],data['Latitude'],c = data['continent'], cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
```
