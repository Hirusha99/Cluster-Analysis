Use the Elbow rule to determine how many clusters are there.


## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set

from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
data = pd.read_csv('/content/014 iris-dataset.csv')
```

```python
data.head()
```

```python
data.describe()
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal.

<i> Use the 'sepal_length' and 'sepal_width' variables.</i>

```python
plt.scatter(data['sepal_length'],data['sepal_width'])
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
kmenas = KMeans(2)
```

```python
x = data.copy()
```

```python
kmenas.fit(x)
```

```python
clusters = kmenas.fit_predict(x)
```

```python
data_with_clusters = data.copy()
data_with_clusters['clusters'] = clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters['sepal_length'],data_with_clusters['sepal_width'],c = data_with_clusters['clusters'],cmap = 'rainbow')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data.

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

# Clustering (scaled data)

```python
kmenas_new = KMeans(2)
```

```python
kmenas_new.fit(x_scaled)
```

```python
clusters_new = kmenas_new.fit_predict(x_scaled)
clusters_new
```

```python
data_new = data.copy()
data_new['clusters'] = clusters_new
data_new
```

```python
plt.scatter(data_new['sepal_length'],data_new['sepal_width'],c=data_new['clusters'] , cmap = 'rainbow')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []
```

### The Elbow Method

```python


for i in range(1,10):
  kmeans = KMeans(i)
  kmeans.fit(x_scaled)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)
```

```python
wcss
```

```python
no_clusters = range(1,10)
plt.plot(no_clusters,wcss)
plt.xlabel('no of clusters')
plt.ylabel('wcss')
```

How many clusters are there?

```python
kmenas_new1 = KMeans(3)
kmenas_new1.fit(x_scaled)
clusters_new1 = kmenas_new1.fit_predict(x_scaled)
data_new1 = data.copy()
data_new1['clusters'] = clusters_new1
plt.scatter(data_new1['sepal_length'],data_new1['sepal_width'],c=data_new1['clusters'] , cmap = 'rainbow')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
kmenas_new1 = KMeans(5)
kmenas_new1.fit(x_scaled)
clusters_new1 = kmenas_new1.fit_predict(x_scaled)
data_new1 = data.copy()
data_new1['clusters'] = clusters_new1
plt.scatter(data_new1['sepal_length'],data_new1['sepal_width'],c=data_new1['clusters'] , cmap = 'rainbow')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```
